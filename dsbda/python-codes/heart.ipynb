```python
''' 
Question : Perform the following operation using python language on the "Air Quality/Heart disease "
            - data cleaning 
            - data integration 
            - data transformation
            - data model building 

''' 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
d_heart = pd.read_csv("/content/heart.csv")
```

```python
d_heart
```

```python
d_heart.isnull().sum()
```

```python
d_heart.duplicated().sum()
```

```python
d_heart.drop_duplicates(inplace = True)
d_heart
```

```python
d_heart.duplicated().sum()
```

```python
columns = ["age","sex","cp","thal"]
df1 = d_heart[columns].loc[0:15]
df1
```

```python
df2 = d_heart[columns].loc[16:30]
df2
```

```python
merging = pd.merge(df1,df2,on="age",how="inner")
merging
```

```python
d_heart['target'] = d_heart['target'].apply(lambda x:1 if x>0 else 0 )
d_heart
```

```python
from sklearn.model_selection import train_test_split

x = merging.drop(["age"],axis=1)
x
```

```python
y = merging["thal_y"]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3) 
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
print(classification_report(y_test,y_pred))
```
