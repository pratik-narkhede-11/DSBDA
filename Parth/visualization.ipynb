```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart.csv')
df.head()
```

```python
df.columns
```

```python
sns.boxplot(x='target',y='age',data=df)
```

```python
plt.hist(df['age'])
plt.xlabel('target')
plt.ylabel('y_target')
plt.title("Histogram")
plt.show(bins=6)
```

```python
plt.plot(df['chol'],df['age'])
plt.xlabel("X_asix")
plt.ylabel("y_axix")
plt.title("Multigraph")
plt.show()
```

```python
sns.lineplot(data=df,x='age',y='target')
```

```python
sns.regplot(x='age',y='sex',data=df)
```
