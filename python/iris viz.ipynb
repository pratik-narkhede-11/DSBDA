```python
sns.boxplot(x=df["sepal_length"])
```

```python
plt.figure(figsize=(16,9))
sns.boxplot(data=df, x="sepal_length", y="sepal_width")
```

```python
plt.figure(figsize=(16,9))
sns.histplot(data=df,x='sepal_length')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(x='sepal_length',y='sepal_width',data=df)
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.lineplot(data=df[['sepal_length','sepal_width','petal_length','petal_width']],palette=['red', 'blue', 'purple', 'pink'])
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.countplot(x='sepal_length',data=df)
plt.show()
```

```python
plt.scatter(x=df.sepal_length[df.species=='setosa'], y=df.sepal_width[(df.species=='setosa')], c="red")
plt.scatter(x=df.sepal_length[df.species=='virginica'], y=df.sepal_width[(df.species=='virginica')])
plt.legend(["Setosa", "Virginica"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
pd.crosstab(df.sepal_length,df.sepal_width).plot(kind="bar",figsize=(20,6))
plt.title('Sepal Length and sepal width ratio')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.savefig('Sepal_length_width.png')
plt.show()
```

```python
pd.crosstab(df.petal_length,df.petal_width).plot(kind="bar",figsize=(15,6))
plt.title('Petal length and petal width')
plt.xlabel('The petal length')
plt.xticks(rotation = 0)
plt.ylabel('The petal width')
plt.show()
```
