```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
#Fetching reviews n comments
reviews = soup.find_all('div' ,{'class' : 't-ZTKy'});
for review in reviews:
    print(review.get_text()+ "\n")
```

```python
# Average Overall rating
avg_rating = soup.find('div',{'class': '_2d4LTz'}).get_text()
print(avg_rating)
```

```python
# Individual ratings
individual_ratings = soup.find_all('div',{'class': '_3LWZlK _1BLPMq'})
for i in individual_ratings:
    print(i.get_text()+"\n")
```

```python
# Fetching tags
tags = soup.find('span',{'class':'B_NuCI'}).get_text()
tags
```

```python
# Fetching customer names
customer_names=soup.find_all('p',{'class':'_2sc7ZR _2V5EHH'})
for j in customer_names:
    print(j.get_text()+"\n")
```
